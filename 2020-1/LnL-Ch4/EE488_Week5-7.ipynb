### Week5-7. Stomer's method/Verlet integration: 비감쇠운동 (Demo. 4-15)

```python
from numpy import zeros, linspace
from math import pi
import matplotlib.pyplot as plt

def osc_2nd_order(U_0, omega, dt, T):
    """
    Solver u'' + omega**2*u = 0 for t in (0,T], u(0)=U_0 and u'(0)=0,
    by a central finite difference method with time step dt.
    """
    dt = float(dt)
    Nt = int(round(T/dt))
    u = zeros(Nt+1)
    t = linspace(0, Nt*dt, Nt+1)
    
    u[0] = U_0
    u[1] = u[0] - 0.5*dt**2*omega**2*u[0]              #using u'(0)=0, can show u[-1]=u[1]
    for n in range(1, Nt):
        u[n+1] = 2*u[n] - u[n-1] - dt**2*omega**2*u[n]
    return u, t

def undamped_linear():
    b = 0.3
    f = lambda v: b*v
    s = lambda u: k*u
    F = lambda t: 0
    
    omega = 1
    m = 1
    k = 1
    U_0 = 1
    V_0 = 0
    
    T = 12*pi
    dt = T/5000.
    
    u, t = osc_2nd_order(U_0, omega, dt, T)

    index = int(len(u)*100/100.)
    plt.plot(t[-index:], u[-index:], 'b-')
    plt.show()

if __name__ == '__main__':
    undamped_linear()
```
