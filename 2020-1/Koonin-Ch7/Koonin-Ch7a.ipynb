```python
import matplotlib.pyplot as plt
import cmath, math
import numpy as np
import os
from matplotlib.animation import FuncAnimation

NSTEP = input("Enter total number of domain steps:  ")  # Total number of domain steps 
NSTEP = int(NSTEP)
H = np.float32(1.0/NSTEP)             # Spacing
phi = np.zeros(NSTEP)     # Space of PHI
diff = np.zeros(NSTEP)    # Space of Error
exactsol = np.zeros(NSTEP)

print("number of steps and H are", NSTEP, H)

### Function Arrangement
def gauss(x,t):
    return math.exp(-np.float32(20.0) * (x-np.float32(0.5))**np.float32(2)/(np.float32(1.0)+np.float32(80)*t))/np.float32(math.sqrt(np.float32(1.0)+np.float32(80)*t))

def exact(x,t):
    return np.float32(gauss(x,t) - gauss(x-np.float32(1.0),t) - gauss(x+np.float32(1.0),t))

### Calculation
time = input("Enter total time: ")
print("We recommend a time step as 0.5*h^2 such as %8.6f" % (0.5*(H**2)))
dt = input("Enter time step (0 to stop): ")
print("time step and time are:", dt, time)

dt = np.float32(dt); time = np.float32(time)

if dt == 0:
    print("dt is zero and program is stopping")
    quit()

NITER=int(time/dt)
dth=dt/H**np.float32(2)
t=np.float32(0); phi[0]=np.float32(0) ; phi[NSTEP-1]=np.float32(0)

for i in range(0, NSTEP-1):
    phi[i]=exact(np.float32(i)*H, t)

for i in range(0,NITER+1):
    pold=0
    for j in range(0, NSTEP-1):
        pnew=phi[j] + dth*(pold + phi[j+1] - np.float32(2)*phi[j])
        pold=phi[j]
        phi[j]=pnew
        exactsol[j] = exact(np.float32(j)*H,time)
    if i % 2 == 1:
        plt.clf()
        fig = plt.figure(figsize=(8,5))
        plot1 = plt.plot(range(0,NSTEP), phi, label="Numerical at iteration = %f" %((i+1)*dt))
        plot2 = plt.plot(range(0,NSTEP), exactsol, label="Analytical")
        plt.ylim(0,4)
        plt.legend(loc=2)
        plt.show()
```
