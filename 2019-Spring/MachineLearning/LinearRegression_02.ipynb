```python
from sklearn import linear_model
import numpy as np

# 1. Classifier
clf = linear_model.BayesianRidge()

# 2. Training data: y=x with noise
X = []
Y = []

for x in np.linspace(0, 10, 101):
    noise = np.random.randint(-50, 50, 1)[0]/200.
    X.append([x+noise])
    noise = np.random.randint(-50, 50, 1)[0]/200.
    Y.append([np.sin(x+noise)])

# 3. Model construction
clf.fit(X, Y)

# Prediction
x_new = [[1.5], [3.5], [5.5], [7.5], [9.5]]
y_new = clf.predict(x_new)

print ("x_new =", x_new)
print ("y_new =", y_new)

# 4. Plot
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(5,5))
fig1 = fig.add_subplot(111)
#fig1.axis([0., 10., 0., 10.])

# Exact y=sin(x)
fig1.plot(np.linspace(0,10,21), 
          np.sin(np.linspace(0,10,21)), 'k-', label='Exact')

# Learning data with noise
fig1.scatter(X, Y, color='k', label='Law data')

# Prediction
fig1.scatter(x_new, y_new, color='r', label='Predict')
plt.show()
```
