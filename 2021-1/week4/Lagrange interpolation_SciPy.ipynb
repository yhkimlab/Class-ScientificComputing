### Lagrange interpolation: SciPy
* SciPy code: https://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.lagrange.html

```python
import numpy as np
from scipy.interpolate import lagrange
import matplotlib.pyplot as plt
from numpy.polynomial.polynomial import Polynomial
x = np.array([0, 1, 2, 3, 4, 5])
y = 1/(25*x**2+1)
poly = lagrange(x, y)

print(Polynomial(poly).coef)

x = np.linspace(0,1,100)
fig,ax = plt.subplots()
plt.plot(x,-0.00727772*x**5+ 0.10987525*x**4-0.62895757*x**3+1.69340061*x**2-2.12857905*x+1)
plt.plot(x, 1/(25*x**2+1))
plt.show()
```
